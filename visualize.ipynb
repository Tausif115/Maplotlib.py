```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

###Basic graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y, label='2x', color='r', linewidth=2, marker='.', markersize=10, markeredgecolor='b', linestyle='--')
plt.title("Our first graph!", fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X axis", fontdict={'fontname':'Comic Sans MS'})
plt.ylabel("Y axis", fontdict={'fontname':'Comic Sans MS'})
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
plt.legend()


plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize
plt.figure(figsize=(5,3), dpi=100)

#fmt = [color][marker][line]
plt.plot(x, y, 'b^--', label = '2x')
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')
#print(x2)

plt.title("Our first graph!", fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X axis", fontdict={'fontname':'Comic Sans MS'})
plt.ylabel("Y axis", fontdict={'fontname':'Comic Sans MS'})
plt.xticks([0,1,2,3])
#plt.yticks([0,2,4,6,8,10])
plt.legend()


plt.show()
```

###Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]
plt.figure(figsize=(6,4))

bars = plt.bar(labels, values)
pattern = ['/','o','*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')

print(gas)
plt.figure(figsize=(8,5))
plt.title('Gas Price $')
#plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
#plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
#plt.plot(gas.Year, gas['South Korea'],'g.-', label='South Korea')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], '.-', label=country)

plt.xlabel('Year', fontdict={'fontname': 'Comic Sans MS', 'fontsize':12})
plt.ylabel('Gas_price', fontdict={'fontname': 'Comic Sans MS', 'fontsize':12})
plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
```
